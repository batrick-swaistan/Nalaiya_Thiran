```python
import matplotlib.pyplot as plt
```

```python
#x-axis values
x=[6,9,8,7,5]
#y-axis
y=[5,4,3,2,1]

plt.plot(x,y)
```

```python
#bar-plot

x=[6,9,8,7,5]

y=[5,4,3,2,1]

plt.bar(x,y)
```

```python
#Histogram(frequency)

y=[5,6,6,6,7,8,9,9]
plt.hist(y)
```

```python
#scatter-plot (relationship between two variables)
x=[6,9,8,7,5]

y=[5,4,3,2,1]

plt.scatter(x,y)
```

```python
#line-chart
x=[6,7,8,9]

y=[9,8,7,6]

plt.plot(x,y)

plt.title('line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
